```python
 face_cascade=cv2.CascadeClassifier(r"E:\ml\Computer-Vision-with-Python\DATA\haarcascades\haarcascade_frontalface_default.xml")
```

```python
def adj_detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade=cv2.CascadeClassifier(r"E:\ml\Computer-Vision-with-Python\DATA\haarcascades\haarcascade_eye.xml")
```

```python
def adj_detect_eyes(img):
    eye_img=img.copy()
    eye_rects=eye_cascade.detectMultiScale(eye_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(eye_img,(x,y),(x+w,y+h),(255,255,255),10)
    return eye_img
```

```python
result=adj_detect_eyes(nadia)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result=adj_detect_eyes(denis)
```
